```python
class NhanVien:
    def __init__(self,id:str="",name:str=""): # contructor
        self.id = id # thuoc tinh Attribute
        self.name = name 
        print("ID:",self.id,"Name:",self.name)

    def show(self): # Phương thức (Method)
        print("ID:",self.id)
        print("Name:", self.name)

    @staticmethod
    def des():
        print("Lớp miêu tả đối tượng nhân viên")

    @classmethod
    def update(cls,id,name):
        return cls(id,name)

nv1  = NhanVien()
nv1.update("2","Nam")
# print(nv1.name)
nv1.show()
```

```python
# Kế thừa 
class NhanVien:
    def __init__(self,id:str="",name:str="",age:int=15): # contructor
        self.id = id # thuoc tinh Attribute
        self.name = name 
        self.age = age
    def show(shelf):
        return f"{self.id}\t{self.name}\t{self.age}"
    
class TruongPhong(NhanVien):
    def __init__(self,id:str="",name:str="",age:int=15,job:str=""):
        NhanVien.__init__(self,id,name,age)
        self.job = job

    def show(self):
        return f"{self.id}\t{self.name}\t{self.age}\t{self.job}"
    
class GiamDoc(TruongPhong):
    def __init__(self,id:str="",name:str="",age:int=15,job:str="",luong:int=""):
        TruongPhong.__init__(self,id,name,age,job)
        self.luong = luong
    def show(self):
        return f"{self.id}\t{self.name}\t{self.age}\t{self.job}\t{self.luong}"
    

 # show da hinh   
tp = TruongPhong("22","C",23,job="CongNghe")
print(tp.show())

gd = GiamDoc("30","D",30,job="DT",luong="2000$")
print(gd.show())
```

```python
# Bài 1: Lớp con kế thừa lớp cha
# Mô tả: Tạo một lớp cha Animal với phương thức speak(). Tạo lớp con Dog kế thừa từ Animal và ghi đè (override) phương thức speak() để trả về âm thanh của chó.
# Yêu cầu:
# Lớp Animal có phương thức speak() trả về "I make a sound".

# Lớp Dog ghi đè phương thức speak() để trả về "Woof!".
# Đầu vào/Đầu ra:
# Gọi Dog().speak() → "Woof!"

# Gọi Animal().speak() → "I make a sound"
```

```python
class DongVat:
    def __init__(self,name:str=""):
        self.name = name

    def speak(self):
        print(f"{self.name}. I make a sound")

class Dog:
    def __init__(self,name:str=""):
        DongVat.__init__(self,name)
        
    def speak(self):
        print(f"{self.name} woof")
    
dv = DongVat("Milo")
dv.speak()

dog = Dog("Hello")
dog.speak()
```

```python
# Bài 2: Kế thừa với thuộc tính
# Mô tả: Tạo lớp cha Vehicle với thuộc tính brand và phương thức info(). Tạo lớp con Car kế thừa từ Vehicle, thêm thuộc tính num_doors và ghi đè phương thức info() để hiển thị thông tin đầy đủ.
# Yêu cầu:
# Lớp Vehicle có thuộc tính brand và phương thức info() trả về "Brand: {brand}".

# Lớp Car có thêm thuộc tính num_doors và ghi đè info() trả về "Brand: {brand}, Number of doors: {num_doors}".
# Đầu vào/Đầu ra:
# Tạo Car("Toyota", 4).info() → "Brand: Toyota, Number of doors: 4"
```

```python
class Vehicle:
    def __init__(self, brand: str):
        self.brand = brand
   
    def info(self):
        print(f"Brand: {self.brand}") 


class Car(Vehicle):
    def __init__(self, brand: str, num_doors: int):
        Vehicle.__init__(self,brand)
        self.num_doors = num_doors


    def info(self):
        return f"Brand: {self.brand}, Number of doors: {self.num_doors}"   
    
xe = Car("Toyota",4)

print(xe.info())        
```

📝 Yêu cầu

🧩 Gợi ý: Sử dụng super(Car, self).get_info() trong get_info() để xử lý kế thừa nhiều.

🧪 Phần kiểm tra

🧠 Nhiệm vụ chính của bạn
Viết lớp Bus kế thừa từ Vehicle với các yêu cầu sau:

Thuộc tính riêng: max_passengers: int

start_engine() in ra: "{brand} {model} bus engine started with max {max_passengers} passengers."

calculate_fuel_efficiency(distance, fuel_used) với hiệu suất giảm theo công thức:

(distance / fuel_used) * (1 - min(max_passengers, 100) / 500)
maintenance_check() in ra: "Checking passenger seating and brakes for {brand} {model}."

get_info() bổ sung thêm số lượng hành khách

Tạo một đối tượng Bus và thêm vào danh sách vehicles để kiểm tra trong test_vehicle_system()

```python
# Lớp phương tiện
class Vehicle: 
    def __init__(self,brand:str = "", model:str="", year:int=0, weight:float=0.0,is_running:bool=False):
        self.brand = brand # thương hiệu
        self.model = model # mẫu
        self.year = year  # năm 
        self. weight =weight # trọng lượng
        self.is_running= is_running # chạy      
    
    def get_info(self):
        return (f"Thương hiệu {self.brand} \t Mẫu {self.model} \t  Năm {self.year} \t Trọng lượng {self.weight}")


# Lớp xe oto
class Car(Vehicle):
    def __init__(self,brand:str = "", model:str="", year:int=0, weight:float=0.0,is_running:bool=False,num_doors:int=0,fuel_type:str=""):
        Vehicle.__init__(self,brand,model,year,weight,is_running)
        self.num_doors = num_doors # so cua
        self.fuel_type= fuel_type # loai nhien lieu

    def start_engine(self):
        self.is_running = True
        print (f"{self.brand} {self.model} Đang sử dụng {self.fuel_type}." )  

    def calculate_fuel_efficiency(self,distance, fuel_used):
        efficiency = 0 
        efficiency = distance / fuel_used # Hieusuat = Quangduong/Nhienlieudadung
        print(f"Hiệu suất = {efficiency}")

    def maintenance_check(self): #kiểm tra dầu 
        print(f"Kiểm tra dầu máy và lốp xe cho {self.brand} {self.model}.")    


# Lớp xe máy
class Motorcycle(Vehicle):
    def __init__(self,brand:str = "", model:str="", year:int=0, weight:float=0.0,is_running:bool=False,has_sidecar:bool=False):
        Vehicle.__init__(self,brand,model,year,weight,is_running)
        self.has_sidecar = has_sidecar #thung xe

    def start_engine(self):
        self.is_running = True
        print(f"{self.brand} {self.model} đã khởi động.")

    def calculate_fuel_efficiency(self,distance, fuel_used):
        efficiency = 0 
        if self.has_sidecar == True:
            efficiency = (distance / fuel_used) * 0.8 
            print(f"Hiệu suất = {efficiency}")
        elif self.has_sidecar == False:
            efficiency = (distance / fuel_used) * 1 
            print(f"Hiệu suất = {efficiency}")
    
    def maintenance_check(self):
        print(f"Kiểm tra dầu, phanh và lốp cho {self.brand} {self.model}.")

# Lớp xe tải 
class Truck(Vehicle):
    def __init__(self,brand:str ="", model:str="", year:int=0, weight:float=0.0,is_running:bool=False,cargo_capacity:int=0):
        Vehicle.__init__(self,brand,model,year,weight,is_running)
        self.cargo_capacity = cargo_capacity # suc chua hang
    
    def start_engine(self):
        self.is_running = True
        print(f"{self.brand} {self.model} đã khởi động.")

    def calculate_fuel_efficiency(self,distance, fuel_used):
        efficiency = 0 
        efficiency = (distance / fuel_used) / (1+(self.cargo_capacity)/1000) 
        print(f"Hiệu suất = {efficiency}")
    
    def maintenance_check(self):
        print(f"Kiểm tra dầu, phanh và lốp cho {self.brand} {self.model}.")

# Lớp xe điện
class ElectricCar(Car):
    def __init__(self,brand:str = "", model:str="", year:int=0, weight:float=0.0,is_running:bool=False,num_doors:int=0,fuel_type:str="",battery_capacity:float=0.0,battery_level:float=0.0):
        Car.__init__(self,brand,model,year,weight,is_running,num_doors,fuel_type)
        self.battery_capacity = battery_capacity # dung luong bin 
        self.battery_level = battery_level # muc bin hien tai

    def charge_battery(self, charge_amount):
        if charge_amount <= 0:
            print("Số kWh sạc phải lớn hơn 0.")
            return
        old_level = self.battery_level
        self.battery_level += charge_amount
        if self.battery_level > self.battery_capacity:
            self.battery_level = self.battery_capacity
        print(f"Đã sạc thêm {charge_amount} kWh (từ {old_level} → {self.battery_level}/{self.battery_capacity} kWh).")


    def start_engine(self):
        self.is_running = True
        print(f"{self.brand} {self.model} đã khởi động.")

    def calculate_fuel_efficiency(self, distance, energy_used):
        if energy_used <= 0:
            print("Lượng điện tiêu thụ phải lớn hơn 0.")
            return None
        efficiency = distance / energy_used
        print(f"Hiệu suất = {efficiency:.2f} km/kWh")
        return efficiency
    
    def maintenance_check(self):
        print(f"Kiểm tra bình điện, phanh và lốp cho {self.brand} {self.model}.")

# Lớp xe Lai 
class HybridCar(ElectricCar):
    def __init__(self,brand:str = "", model:str="", year:int=0, weight:float=0.0,is_running:bool=False,num_doors:int=0,fuel_type:str="",battery_capacity:float=0.0,battery_level:float=0.0):
        ElectricCar.__init__(self,brand,model,year,weight,is_running,num_doors,fuel_type,battery_capacity,battery_level)

    def start_engine(self):
        self.is_running = True      
        print(f"{self.brand} {self.model} đã khởi động.")

    def mode(self,distance, fuel_used,energy_used):
        if self.fuel_type == "Petrol":
            efficiency = (distance / fuel_used) * 0.9
            print(f"Hiệu suất = {efficiency}")
        elif self.fuel_type == "Electric":
            efficiency = (distance / energy_used)
            print(f"Hiệu suất = {efficiency}")
                  
    def maintenance_check(self):
        print(f"Kiểm tra dầu, phanh và lốp cho {self.brand} {self.model}.")       


def test_vehicle_system(vehicles):

    for i in vehicles:
        print("\n")
        print(i.get_info())
        print(f"Trạng thái: {'Đang bật' if i.is_running else 'Đang tắt'}")
        i.start_engine()
        efficiency = None
        if isinstance(i, HybridCar):
            efficiency = i.calculate_fuel_efficiency(400, 80)  #
        elif isinstance(i, ElectricCar):
            efficiency = i.calculate_fuel_efficiency(150, 25)  #  6 km/kWh
        elif isinstance(i, Car):
            efficiency = i.calculate_fuel_efficiency(600, 40)  #  15 km/l
        elif isinstance(i, Motorcycle):
            efficiency = i.calculate_fuel_efficiency(300, 8)   #  37.5 km/l
        elif isinstance(i, Truck):
            efficiency = i.calculate_fuel_efficiency(400, 80)  #  5 km/l


        i.maintenance_check()

        if hasattr(i, "charge_battery"):
            i.charge_battery(10)



vehicles = [
    Car("Toyota", "Camry", 2020, 1500, True, 4, "Petrol"),
    # Car("Tesla", "Model 3", 2021, 1600, False, 4, "Electric"),
    # Motorcycle("Yamaha", "PG 1", 2015,1000,True,True),
    # Truck("Ford", "F-150", 2019, 3000, False, 5000),
    ElectricCar("Tesla", "Model 3", 2021, 1600, False, 4, "Electric", 75, 50),
    HybridCar("Toyota", "Prius", 2022, 1400, False, 4, "Electric", 50, 20,),
    # HybridCar("Toyota", "Prius", 2022, 1400, False, 4, "Petrol", 50, 20,)
]

test_vehicle_system(vehicles)
```
