```python
def read_file_txt(path_file:str):
    with open(path_file,mode="r",encoding="utf-8") as f:
        # content = f.read()
        result = []
        content = f.readlines()
        for item in content :
            if item == '\n':
                continue
            else: 
                result.append(item.replace("\n","").strip())
        return result

read_file_txt(r"C:\Users\Admin\Downloads\Hoc Python\KhoaHocT3H\Test_text.txt")
```

```python
def write_file_txt(path_file:str, content:list):
    # with open(path_file,mode='w',encoding="utf_8") as f:
    #     f.write("\n".join(content))
    with open(path_file,mode="w",encoding="utf-8") as f:
        f.write("\n".join(content))
data = ["RUN python main.py","bbbbbbbb"]
write_file_txt(r"C:\Users\Admin\Downloads\Hoc Python\KhoaHocT3H\Test_text.txt",data)
```

```python
import json

def read_file_json(path_file:str) -> list:
    with open(path_file, mode= "r",encoding="utf-8") as f:
        data = json.load(f)
        print(data[0]["model"])
    return data 
read_file_json(r"C:\Users\Admin\Downloads\mtcars-parquet.json")
```

```python
def wirte_file_json(path_file:str, content:list):
    with open(path_file,mode="a",encoding="utf-8")as f:
        json.dump(content,f,indent=2,sort_keys=True)

data = [{"docker":"Run python main.py"}]
wirte_file_json(r"C:\Users\Admin\Downloads\mtcars-parquet.json",data)
```
